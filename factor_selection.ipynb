# ETF(팩터) 수집

```python
from pykrx import stock
from pykrx import bond
```

```python
df_f = pd.DataFrame()
tickers = ['174350','147970','227570','123310','226980','244670','244660','211900','210780']
for ticker in tickers:
    df = stock.get_etf_ohlcv_by_date("20160701", "20231231", str(ticker))['NAV']
    df_f = pd.concat([df_f, df], axis = 1)
    df_f.rename(columns = {'NAV' : str(ticker)})
    
name = ['타이거로우볼','타이거모멘텀','타이거우량가치','타이거코스피인버스','코덱스코스피중소형주','코덱스벨류','코덱스퀄리티',
        '코덱스배당성장','타이거고배당']
df_f.columns = name
df_f
```

```python
start_date = datetime(2016,7,1)
end_date = datetime(2023,12,31)

date_list = []

current_date = start_date
while current_date <= end_date:
    date_list.append(str(current_date.strftime('%Y-%m')))
    #넣으면 다음 달로 이동
    if current_date.month == 12:
        current_date = current_date.replace(year = current_date.year + 1, month = 1)
    else:
        current_date = current_date.replace(month = current_date.month + 1)
print(date_list)
```

```python
#팩터의 월별 수익률을 계산하는 함수
def date_to_month(df_f: pd.DataFrame):

    col = df_f.columns.tolist()
    df_ret = pd.DataFrame(columns = col)

    for date in date_list:
        df_m = df_f[pd.to_datetime(df_f.index).strftime("%Y-%m") == date]
        if not df_m.empty:  # df_m이 비어 있지 않은 경우에만 계산 수행
            ret_list_index = [df_m.index[0]]  # 첫 번째 날짜를 인덱스로 사용
            ret_list = (df_m[-1:].values - df_m[:1].values) / df_m[:1].values
            ret_list = pd.DataFrame(ret_list, columns = col, index = ret_list_index)  # 'date' 컬럼 제외하고 나머지 사용
            df_ret = pd.concat([df_ret, ret_list], axis = 0)
    return df_ret
df_ret = date_to_month(df_f)
```

```python
#index를 '년-월'로 통일
df_ret.index = df_ret.index.strftime('%Y-%m')
df_factor = df_ret
```

# 국지 팩터 선정

```python
# 경제변수를 4등분하여 4개 구간을 만듦
# 경제변수 1 기준 4개 구간인 1-1, 1-2, 1-3, 1-4가 생성됨
# 경제변수 1의 각 4개 구간에서 팩터의 평균 수익률들 중 최대 수익률을 [팩터, 경제변수] 잠재최대수익률 설정
# 즉, 잠재 최대 수익률은 (팩터의 수) x (경제변수의 수) 만큼 생성됨
# 잠재 최대 수익률을 팩터 기준으로 묶어서, 그 최대치를 [팩터] 잠재최대수익률로 설정

# input : 팩터(수익률) 묶음, 경제변수 묶음
# output : 각 팩터 별 잠재최대수익률
```

```python
def potential_maximum_return(features: pd.DataFrame, factors: pd.DataFrame):
    #각 feature 별 시계열, 수익률, 구간으로 이루어진 DataFrame을 형성하고 dict에 저장
    features = features.astype(float)
    feature_list = features.columns.to_list()
    factor_list = factors.columns.to_list()
    labels = ['very low', 'low', 'high', 'very high']
    dic = dict()
    for feature in feature_list:
        df_labeling = pd.qcut(features[feature], q = 4, labels = labels)
        df_fe = pd.concat([features[feature], df_labeling], axis = 1)
        df_fe.columns = [feature, 'condition']
        #딕셔너리 생성
        
        for factor in factor_list: #팩터 명(반복문)
            df_new = pd.concat([factors[factor], df_fe['condition']], axis = 1)
            df_new.columns = [factor, 'condition']
            dic[feature + '/' + factor] = df_new.groupby('condition').mean().max()
    return dic
```

```python
#경제변수 / 팩터
potential_maximum_return(df_feature, df_factor)
```

```python
#팩터 기준으로 최종 팩터별 잠재 수익률 뽑기
potential_ret = potential_maximum_return(df_feature, df_factor)
potential_ret.keys()
```

```python
#팩터 기준으로 최종 팩터별 잠재 수익률 뽑기
from collections import defaultdict
dic = defaultdict(int)

for factor in df_factor.columns:
    dic[factor] = []

for s in potential_ret.keys():
    factor_name = s.split('/')[1]
    if factor_name in dic.keys():
        dic[factor_name].append(potential_ret[s][0])

for key, value in dic.items():
    dic[key] = max(dic[key])
    
dic
```

```python
df_final = pd.DataFrame(dic.values(), columns = ['잠재최대수익률'], index = dic.keys()).sort_values(by = '잠재최대수익률', ascending = False)
df_final
```

```python
result = ((df_final['잠재최대수익률'] + 1)**12 - 1) * 100
```
