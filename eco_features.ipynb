```python
import pandas as pd
import numpy as np
import pandas_datareader as pdr
import statsmodels.api as sm
from datetime import datetime
import yfinance as yf
import os

os.getcwd()
```

```python
# 방향성 계산 함수

def calculate_directional_indicator(data, window):
    x = np.arange(window)
    x = sm.add_constant(x)
    directional_indicator = []
        
    for i in range(window, len(data)):
        y = data[i - window:i]
        model = sm.OLS(y, x)
        results = model.fit()
    
        slope = results.params[1]
        mean_absolute_value = np.mean(np.abs(y))
    
        directional_indicator.append(slope / mean_absolute_value)

    return directional_indicator
```

```python
start = datetime(2016, 1, 1)
end = datetime.now().date()
kospi = pdr.DataReader('kospi', 'naver', start, end).astype(float)['Close']

kospi_20 = calculate_directional_indicator(kospi,20)

#코스피
eco_data = pd.DataFrame({'kospi': kospi})
eco_data['kospi_20'] = np.nan
eco_data['kospi_20'][20:] = kospi_20

kospi_60 = calculate_directional_indicator(kospi,60)
eco_data['kospi_60'] = np.nan
eco_data['kospi_60'][60:] = kospi_60

#원/달러
w_d = yf.download('USDKRW=X', start, end).astype(float)['Close']
wd_20 = calculate_directional_indicator(w_d,20)
wd_60 = calculate_directional_indicator(w_d,60)

w_d = w_d.to_frame()

w_d['wd_20'] = np.nan
w_d['wd_20'][20:] = wd_20

w_d['wd_60'] = np.nan
w_d['wd_60'][60:] = wd_60

#원/엔
w_y= yf.download('KRWJPY=X', start, end).astype(float)['Close']
wy_20 = calculate_directional_indicator(w_y,20)
wy_60 = calculate_directional_indicator(w_y,60)

w_y = w_y.to_frame()

w_y['wy_20'] = np.nan
w_y['wy_20'][20:] = wy_20

w_y['wy_60'] = np.nan
w_y['wy_60'][60:] = wy_60

#코스피변동성
vkospi = pd.read_csv('KOSPI Volatility 과거 데이터.csv')
vkospi.set_index(vkospi.columns[0], inplace=True)
vkospi = vkospi.iloc[::-1]
vkospi = vkospi['종가']

vkospi_20 = calculate_directional_indicator(vkospi,20)
vkospi_60 = calculate_directional_indicator(vkospi,60)

vkospi = vkospi.to_frame()

vkospi['vkospi_20'] = np.nan
vkospi['vkospi_20'][20:] = vkospi_20

vkospi['vkospi_60'] = np.nan
vkospi['vkospi_60'][60:] = vkospi_60

#wti
wti = pd.read_csv('WTI유 선물 과거 데이터.csv')
wti.set_index(wti.columns[0], inplace=True)
wti = wti.iloc[::-1]
wti = wti['종가']

wti = wti.shift()

wti_20 = calculate_directional_indicator(wti,20)
wti_60 = calculate_directional_indicator(wti,60)

wti = wti.to_frame()

wti['wti_20'] = np.nan
wti['wti_20'][20:] = wti_20

wti['wti_60'] = np.nan
wti['wti_60'][60:] = wti_60

#미국채10년금리
us_10 = pd.read_csv('미국 10년물 국채 금리  채권수익률 과거 데이터.csv')
us_10.set_index(us_10.columns[0], inplace=True)
us_10 = us_10.iloc[::-1]
us_10 = us_10['종가']

us_10= us_10.shift()

us_10_20 = calculate_directional_indicator(us_10,20)
us_10_60 = calculate_directional_indicator(us_10,60)

us_10= us_10.to_frame()

us_10['us_10_20'] = np.nan
us_10['us_10_20'][20:] = us_10_20

us_10['us_10_60'] = np.nan
us_10['us_10_60'][60:] = us_10_60

#미국채2년금리
us_2 = pd.read_csv('미국 2년 채권수익률 과거 데이터.csv')
us_2.set_index(us_2.columns[0], inplace=True)
us_2 =us_2.iloc[::-1]
us_2 =us_2['종가']
us_2= us_2.shift()

us_2_20 = calculate_directional_indicator(us_2,20)
us_2_60 = calculate_directional_indicator(us_2,60)

us_2= us_2.to_frame()

us_2['us_2_20'] = np.nan
us_2['us_2_20'][20:] = us_2_20

us_2['us_2_60'] = np.nan
us_2['us_2_60'][60:] = us_2_60

#미국채 팁스 금리
us_tips = pd.read_csv('US10YTIP=RR 개요.csv')
us_tips.set_index(us_tips.columns[0], inplace=True)
us_tips =us_tips.iloc[::-1]
us_tips =us_tips['종가']
us_tips= us_tips.shift()

us_tips_20 = calculate_directional_indicator(us_tips,20)
us_tips_60 = calculate_directional_indicator(us_tips,60)

us_tips= us_tips.to_frame()

us_tips['us_tips_20'] = np.nan
us_tips['us_tips_20'][20:] = us_tips_20

us_tips['us_tips_60'] = np.nan
us_tips['us_tips_60'][60:] = us_tips_60

#vix
vix = pd.read_csv('CBOE Volatility Index 과거 데이터.csv')
vix.set_index(vix.columns[0], inplace=True)
vix =vix.iloc[::-1]
vix =vix['종가']
vixs= vix.shift()

vix_20 = calculate_directional_indicator(vix,20)
vix_60 = calculate_directional_indicator(vix,60)

vix= vix.to_frame()

vix['vix_20'] = np.nan
vix['vix_20'][20:] = vix_20

vix['vix_60'] = np.nan
vix['vix_60'][60:] = vix_60
```
